```python
import pandas as pd
import numpy as np
```

```python
# Reading in the data
df = pd.read_csv('datasets/interpreting_results.csv')
```

```python
df.head()
```

```python
# Split the top feature combinations
df['split_var'] = [i.split() for i in df['variable']]
df.head()
```

```python
# Create top value add and subrtractors for the home value
top = df.sort_values(by= 'coefs', ascending = False).head(20)
bottom = df.sort_values(by= 'coefs').head(20)
top.head()
```

```python
# Create a list of unique features

top['split_var_1'] = [i[0] for i in top['split_var']]
top['split_var_2'] = [i[1] for i in top['split_var']]
top.head()
```

```python
# Create a list of unique features

bottom['split_var_1'] = [i[0] for i in bottom['split_var']]
bottom['split_var_2'] = [i[1] for i in bottom['split_var']]
bottom.head()
```

```python
# create and order top value add features
top_d = {}
for i in top['split_var_1']:
    if i in top_d.keys():
        top_d[i] +=1
    else:
        top_d[i] = 1
for i in top['split_var_2']:
    if i in top_d.keys():
        top_d[i] +=1
    else:
        top_d[i] = 1
        
top_df = pd.DataFrame(data = top_d, index=top_d.values())
top_df = top_df.iloc[[0]].T#.columns = ['feature', 'number']
top_df['feature'] = top_df.index
top_df = top_df.reset_index()
top_df = top_df.drop(labels = 'index', axis = 1)
top_df['occurances'] =top_df[2]
top_df = top_df.drop(labels = 2, axis = 1)
top_df.sort_values(by = 'occurances', ascending = False).head(7)
```

```python
bottom_d = {}
for i in bottom['split_var_1']:
    if i in bottom_d.keys():
        bottom_d[i] +=1
    else:
        bottom_d[i] = 1
for i in bottom['split_var_2']:
    if i in bottom_d.keys():
        bottom_d[i] +=1
    else:
        bottom_d[i] = 1
bottom_d = pd.DataFrame(data = bottom_d, index=bottom_d.values())
bottom_d = bottom_d.iloc[[0]].T#.columns = ['feature', 'number']
bottom_d['feature'] = bottom_d.index
bottom_d = bottom_d.reset_index()
bottom_d = bottom_d.drop(labels = 'index', axis = 1)
bottom_d['occurances'] =bottom_d[2]
bottom_d = bottom_d.drop(labels = 2, axis = 1)
bottom_d.sort_values(by = 'occurances', ascending = False)
bottom_d.sort_values(by = 'occurances', ascending = False).head(7)
```
